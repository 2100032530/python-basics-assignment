```python
with open("students.txt","r") as file:
  line1=file.readline()
  line2=file.readline()
print(line1 , end="")
print(line2 , end="")
```

```python
with open("students.txt","r") as file:
  lines=file.readlines()
print(lines)
print(type(lines))
```

```python
for each_line in lines:
  print(each_line)
```
