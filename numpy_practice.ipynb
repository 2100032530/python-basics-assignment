Checking numpy library speed

```python
import time
import numpy as np
size=10000000
numbers=list(range(size))
list_start_time=time.time()
for i,v in enumerate(numbers):
  numbers[i]=v+1
list_end_time=time.time()

numpy_array=np.arange(size)
array_start_time=time.time()
result_array=numpy_array+1
numpy_end_time=time.time()
list_time=list_end_time - list_start_time
array_time=numpy_end_time-array_start_time

print(f"{list_time:.4f} is time taken by list")
print(f"{array_time:.4f} is time taken by array")

```
